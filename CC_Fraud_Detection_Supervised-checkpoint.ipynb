
## Create Supervised Models for Fraud Detection

#### Read CSV Data
##### - source: https://www.kaggle.com/mlg-ulb/creditcardfraud

```python
import pandas as pd
csv_file = 'creditcard.csv'
fraud_df = pd.read_csv(csv_file)
display(fraud_df)
```

```python
import numpy as np
import sklearn as sk
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# get features df
df_features = fraud_df.drop("Class", axis=1)
features_array = np.array(df_features)

# get label df
df_label = fraud_df["Class"]
label_array = np.array(df_label)

# scale features
scaler = StandardScaler()
scaled_features_array = scaler.fit_transform(df_features)
print(scaled_features_array)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# principal component analysis - reduce number of features to X
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(scaled_features_array)

# plot principal components
features = range(pca.n_components_)
plt.figure(figsize=(15, 5))
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('Variance')
plt.xticks(features)
plt.title("Importance of the Principal Components based on inertia")
plt.show()
```

```python
# split into testing/training sets
x_train, x_test, y_train, y_test = train_test_split(df_features, df_label, test_size=.3, random_state=42)

x_train = pca.fit_transform(x_train) # x_train
x_test = pca.transform(x_test) # x_test

# # verify that there are X number of features
print(x_train.shape)
print(x_test.shape)
```

### Isolation Forest Model - Anomaly Detection

```python
from sklearn.ensemble import IsolationForest

# calculate ratio of fraudulent transactions (contamination argument in IsolationForest model)
fraud = fraud_df[fraud_df["Class"] == 1]
not_fraud = fraud_df[fraud_df["Class"] == 0]
outlier_fraction = len(fraud) / float(len(not_fraud))
print(outlier_fraction)

# specify and fit isolation forest model to training set
ifc = IsolationForest(max_samples=len(x_train), contamination=outlier_fraction, random_state=1)
ifc.fit(x_train)

# calculate average anomally score of base classifiers
a_scores_pred = ifc.decision_function(x_train)

# make predictions on test set
pred_df = ifc.predict(x_test)
print(pred_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, matthews_corrcoef
from sklearn import model_selection

# build evaluation matrix on test set
# map predictions to 0s and 1s and find sum of predictions =! to the test set labels
pred_df[pred_df == 1] = 0
pred_df[pred_df == -1] = 1
n_errors = (pred_df != y_test).sum()
print("Number of Errors:" , n_errors)

# visualize confusion matrix
matrix_labels = ['Not Fraud', 'Fraud']
conf_matrix = confusion_matrix(y_test, pred_df)
plt.figure(figsize=(12,12))
sns.heatmap(conf_matrix, xticklabels=matrix_labels, yticklabels=matrix_labels, annot=True, fmt="d");
plt.title("Confusion matrix")
plt.ylabel("True class")
plt.xlabel("Predicted class")
plt.show()
# calculate the false positive rate from the confusion matrix
fp = conf_matrix[0][1]
tn = conf_matrix[1][1]
print("False Positive Rate: {}".format(fp / (fp+tn)))

# evaluate and summary performance metrics
print("The model used: {}".format("Isolation Forest"))
acc = accuracy_score(y_test, pred_df)
print("Accuracy of model: {}".format(acc))
prec = precision_score(y_test, pred_df, zero_division=1)
print("Precision: {}".format(prec))
recall = recall_score(y_test, pred_df)
print("Recall: {}".format(recall))
f1 = f1_score(y_test, pred_df)
print("F1 score: {}".format(f1))
mcc = matthews_corrcoef(y_test,pred_df)
print("Matthews correlation coefficient: {}".format(mcc))
```

### Random Forest - Classification

```python
from sklearn.ensemble import RandomForestClassifier

# specify random forest model and fit to training set
rf = RandomForestClassifier()
rf.fit(x_train, y_train)

# make predictions on test set
pred_df = rf.predict(x_test)
```
