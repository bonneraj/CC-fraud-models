## Create Supervised Models for Fraud Detection

#### Read CSV Data
##### - source: https://www.kaggle.com/mlg-ulb/creditcardfraud

```python
import pandas as pd
csv_file = 'creditcard.csv'
fraud_df = pd.read_csv(csv_file)
display(fraud_df)
```

```python
import numpy as np
import sklearn as sk
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# get features df
df_features = fraud_df.drop("Class", axis=1)
features_array = np.array(df_features)

# get label df
df_label = fraud_df["Class"]
label_array = np.array(df_label)

# scale features
scaler = StandardScaler()
scaled_features_array = scaler.fit_transform(df_features)
print(scaled_features_array)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# principal component analysis - reduce number of features to X
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(scaled_features_array)

# plot principal components
features = range(pca.n_components_)
plt.figure(figsize=(15, 5))
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('Variance')
plt.xticks(features)
plt.title("Importance of the Principal Components based on inertia")
plt.show()
```

```python
# split into testing/training sets
x_train, x_test, y_train, y_test = train_test_split(df_features, df_label, test_size=.3, random_state=42)

x_train = pca.fit_transform(x_train) # x_train
x_test = pca.transform(x_test) # x_test

# # verify that there are X number of features
print(x_train.shape)
print(x_test.shape)
```

```python
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.cluster import KMeans 
from sklearn.metrics import silhouette_score

kmeans = KMeans(
init = "random",
n_clusters = 3,
n_init = 10,
max_iter = 300,
random_state = 42)

kmeans.fit(x_train)
```

```python
# lowest sum of squared error (SSE)
print("lowest sum of squared error (SSE): {}".format(kmeans.inertia_))

# final locations of centroid
print("final locations of centroid: {}".format(kmeans.cluster_centers_))

# number of iterations required to converge
print("number of iterations required to converge: {}".format(kmeans.n_iter_))
```
